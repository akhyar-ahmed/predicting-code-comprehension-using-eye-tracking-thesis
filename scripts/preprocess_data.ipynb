# Preprocessing data
Only output 9/9 code snippets without the additional 10th snippet that only exists for some participants

```python
# load eye-tracking data

import os
import pandas as pd

eyelink_report_dir = "../../../raw_fixation_reports"
directory = os.fsencode(eyelink_report_dir)

# single experiment data frames
dfs = []

# read in all single experiment data
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    dfs.append(pd.read_csv(os.path.join(eyelink_report_dir, filename),sep='\t'))

# concatenate single experiment data to total data frame
df = pd.concat(dfs)

# clean up "." strings into 0.0 floats
df["NEXT_FIX_DISTANCE"] = pd.to_numeric(df["NEXT_FIX_DISTANCE"].replace(".", 0.0)) # replace eye-link data report "." with floats 0.0 and change column d_type to numeric
df["NEXT_FIX_X"] = pd.to_numeric(df["NEXT_FIX_X"].replace(".", 0.0)) # replace eye-link data report "." with floats 0.0 and change column d_type to numeric
df["NEXT_FIX_Y"] = pd.to_numeric(df["NEXT_FIX_Y"].replace(".", 0.0)) # replace eye-link data report "." with floats 0.0 and change column d_type to numeric
df["CURRENT_FIX_PUPIL"] = pd.to_numeric(df["CURRENT_FIX_PUPIL"].replace(".", 0.0))
df["NEXT_SAC_AMPLITUDE"] = pd.to_numeric(df["NEXT_SAC_AMPLITUDE"].replace(".", 0.0))
df["NEXT_SAC_AVG_VELOCITY"] = pd.to_numeric(df["NEXT_SAC_AVG_VELOCITY"].replace(".", 0.0))
df["NEXT_SAC_PEAK_VELOCITY"] = pd.to_numeric(df["NEXT_SAC_PEAK_VELOCITY"].replace(".", 0.0))

# add precision column, precision defines whether the comprehension task was answered correctly (correct = 1)
df["precision"] = df.apply(lambda row: 1 if (row.correct_option == row.KEY_STROKE) else 0, axis=1)

df
```

```python
# load experiment answers
import os
import pandas as pd
answer_report_dir = "../../../experiment_answers"
directory_answers = os.fsencode(answer_report_dir)

# single experiment data frames
dfs_ans = {}

# read in all single experiment data
for file in os.listdir(directory_answers):
    filename = os.fsdecode(file)
    print(filename)
    dfs_ans[filename[:-5]] = pd.read_excel(os.path.join(answer_report_dir, filename),sheet_name="Tabelle1")
    dfs_ans[filename[:-5]] = dfs_ans[filename[:-5]].drop('Session_Name_', axis=1)

# concatenate single experiment data to total data frame
dfs_ans
```

```python
# removing faulty participant's data
#df = df[df["RECORDING_SESSION_LABEL"] != "P205"]
df = df[df["RECORDING_SESSION_LABEL"] != "P001"]
#df = df[df["RECORDING_SESSION_LABEL"] != "P002"]

# keep only 10th code snippet without task
#df = df[df["code_snippet_id"] != "369-1404-V1"]
df = df[df["code_snippet_id"] == "369-1404-V2"]

# only keep fixations on the code snippets (removing fixations on the task & answers)
df = df[df["CURRENT_FIX_X"].between(430, 1400)]

df
```

```python
# remove next fixation distance if it is outside of the code snippet area
df = df[df['NEXT_FIX_X'].between(430, 1400)]
```

```python
# build data frame for machine learning model input
import numpy as np

# feature list builder
MAX_FIXATION_AMOUNT = 1225+1 # max amount of tracking rows per code snippet (considering all code snippets)
feature_list = []
for i in range(MAX_FIXATION_AMOUNT):
    feature_list.extend(
        [
            f"{i}_i",
            f"{i}_dur",
            f"{i}_pos_x",
            f"{i}_pos_y",
            #f"{i}_pupil_size",
            f"{i}_nxt_fix_dis",
            #f"{i}_nxt_fix_dir",
        ]
    )
feature_list.append("precision")
feature_list.append("subj_difficulty")
feature_list.append("code_snippet_id")

# group data by participant
df_grouped_by_participant = df.groupby(["RECORDING_SESSION_LABEL"])

# init data frame with final columns (fixed length, based on max fixation amount)
participant_amount = len(df_grouped_by_participant["RECORDING_SESSION_LABEL"].unique())
code_snippet_amount = 10

d = pd.DataFrame(0, index=np.arange(participant_amount*code_snippet_amount), columns=feature_list + ["RECORDING_SESSION_LABEL"]) # zero array

print(f"d shape: {d.shape}")
# filling the dataframe
i = 0

direction_encoding = {
    ".": 0,
    "DOWN": 1,
    "LEFT": 2,
    "RIGHT": 3,
    "UP": 4
}

subj_rating_encoding = {
    "1": 0.0,
    "2": 0.25,
    "3": 0.5,
    "4": 5.75,
    "5": 1.0,
}

for participant_name, participant_group in df_grouped_by_participant:
    print(f"Looking at {participant_name} with shape: {participant_group.shape}")

    df_grouped_by_participant_cs = participant_group.groupby(["code_snippet_id"])

    for code_snippet_name, participant_code_snippet_group in df_grouped_by_participant_cs:
        print(f"    Looking at {code_snippet_name} with shape: {participant_code_snippet_group.shape}")
        j = 0
        for rj, row in participant_code_snippet_group.iterrows():
            if j == 0:
                d.iloc[i,0] = 0
                d.iloc[i,1] = row["CURRENT_FIX_DURATION"]
                d.iloc[i,2] = row["CURRENT_FIX_X"]
                d.iloc[i,3] = row["CURRENT_FIX_Y"]
                #d.iloc[i,4] = row["CURRENT_FIX_PUPIL"]
                d.iloc[i,4] = row["NEXT_FIX_DISTANCE"]
                #d.iloc[i,6] = direction_encoding[row["NEXT_FIX_DIRECTION"]]
            else:
                d.iloc[i,j*5] = j
                d.iloc[i,j*5+1] = row["CURRENT_FIX_DURATION"]
                d.iloc[i,j*5+2] = row["CURRENT_FIX_X"]
                d.iloc[i,j*5+3] = row["CURRENT_FIX_Y"]
                #d.iloc[i,j*5+4] = row["CURRENT_FIX_PUPIL"]
                d.iloc[i,j*5+4] = row["NEXT_FIX_DISTANCE"]
                #d.iloc[i,j*7+6] = direction_encoding[row["NEXT_FIX_DIRECTION"]]

                #d.iloc[i,j*5+6] = row["NEXT_SAC_AMPLITUDE"]
                #d.iloc[i,j*5+7] = row["NEXT_SAC_AVG_VELOCITY"]
                #d.iloc[i,j*5+8] = row["NEXT_SAC_PEAK_VELOCITY"]
                #d.iloc[i,j*5+9] = 1 if row["NEXT_SAC_CONTAINS_BLINK"] == "true" else 0

            precision = row["precision"]
            label = row["RECORDING_SESSION_LABEL"]
            j += 1

        d.iloc[i,-4] = 1 if dfs_ans[row["RECORDING_SESSION_LABEL"]][dfs_ans[row["RECORDING_SESSION_LABEL"]]['code_snippet_id'] == row["code_snippet_id"]]["CS_UNDERSTANDING"].values[0] == "t" else 0 #precision
        d.iloc[i,-3] = dfs_ans[row["RECORDING_SESSION_LABEL"]][dfs_ans[row["RECORDING_SESSION_LABEL"]]['code_snippet_id'] == row["code_snippet_id"]]["CS_SUBJ_DIFFICULTY"]
        d.iloc[i,-2] = row["code_snippet_id"]
        d.iloc[i,-1] = label
        i += 1

# remove 0 rows
d = d.loc[(d != 0).any(axis=1)]

# normalize subjective difficulty values
def min_max_scaling(series):
    return (series - series.min()) / (series.max() - series.min())

# done manually at line 88: d.iloc[:,-3] = min_max_scaling(d.iloc[:,-3])
d["subj_difficulty"] =  d["subj_difficulty"].replace(1,0.0)
d["subj_difficulty"] =  d["subj_difficulty"].replace(2,0.25)
d["subj_difficulty"] =  d["subj_difficulty"].replace(3,0.5)
d["subj_difficulty"] =  d["subj_difficulty"].replace(4,0.75)
d["subj_difficulty"] =  d["subj_difficulty"].replace(5,1.0)

#d["precision"] =  d["precision"].replace(5,1.0)

# save to file
d.to_csv("../../data_public/final_preprocessing/preprocessed_cs_10_without_task.csv")

d
```
